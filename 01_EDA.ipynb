#### Import Libraries

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
res_data = pd.read_csv("./data/samples_df4.csv", index_col=0)
res_data.head()
```

```python
res_data.info()
```

### Visualization - Memory

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(res_data['time'], res_data['assigned_mem'], label = "Memory Assigned")
ax.plot(res_data['time'], res_data['mem_available'], label = "Memory Available")
ax.plot(res_data['time'], res_data['mem_requested'], label = "Memory Requested")
ax.plot(res_data['time'], res_data['mem_usage'], label = "Memory Usage")

plt.xlim(min(res_data['time']), max(res_data['time']))
plt.legend()
plt.savefig("./visuals/Memory_Overview.png")
plt.show()
```

### Visualization - CPU

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(res_data['time'], res_data['cpu_available'], label = "CPU Available")
ax.plot(res_data['time'], res_data['cpu_requested'], label = "CPU Requested")
ax.plot(res_data['time'], res_data['cpu_usage'], label = "CPU Usage")

plt.xlim(min(res_data['time']), max(res_data['time']))
plt.legend()
plt.savefig("./visuals/CPU_Overview.png")
plt.show()
```
